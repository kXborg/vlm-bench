```python
import torch
from PIL import Image
from transformers import AutoModelForCausalLM, CodeGenTokenizerFast as Tokenizer

model_id = "vikhyatk/moondream1"
model = AutoModelForCausalLM.from_pretrained(model_id, trust_remote_code=True)
tokenizer = Tokenizer.from_pretrained(model_id)
```

# Check Model Size and Params

```python
num_params = sum(p.numel() for p in model.parameters())
print("Parameters:", num_params)

dtype = next(model.parameters()).dtype
bytes_per_param = {torch.float32: 4, torch.float16: 2, torch.bfloat16: 2}.get(dtype, 4)
size_gb = num_params * bytes_per_param / (1024**3)
print(f"Approx size in memory: {size_gb:.2f} GB")
print('DType: ', dtype)
```

```python
image = Image.open('../../tasks/bird.jpg')
enc_image = model.encode_image(image)
print(model.answer_question(enc_image, "What is this image about?", tokenizer))
```
